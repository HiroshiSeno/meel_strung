## 2.3.1
$Au = b$が１個の未知数$u$に関する$m$個の方程式$a_iu = b_i$であるとする．平方和$E(u) = (a_1u-b_1)^2+\ldots+(a_mu-b_m)^2$に対して，最小化する$\hat{u}$を微積分によって求めよ．次に，1列のみの$A$に関する式$A^TA\hat{u} = A^Tb$について線形代数を用いて，同じ$\hat{u}$に到達せよ．

### 解答

$E(u) = (a_1u-b_1)^2+\ldots+(a_mu-b_m)^2$を$u$で微分する．最小化する点では$\partial{E(u)}/\partial{u}=0$であるため，

$$
\begin{align}
\frac{\partial{E(u)}}{\partial{u}} = 2a_1(a_1\hat{u}-b_1) + 2a_2(a_2\hat{u}-b_2) +\ldots + 2a_m(a_m\hat{u}-b_m) &= 0 \\
(a_1^2+a_2^2+\ldots+a_m^2)\hat{u} - (a_1b_1+a_2b_2+\ldots+a_mb_m) &= 0
\end{align}
$$

ゆえに以上を計算すると，
$$
\hat{u} = \frac{a_1b_1+a_2b_2+\ldots+a_mb_m}{a_1^2+a_2^2+\ldots+a_m^2}
$$

また$A^TA\hat{u} = A^Tb$を解いてみる．

$$
\begin{align}
\begin{pmatrix}a_1 & a_2 & \ldots & a_m\end{pmatrix}\begin{pmatrix}a_1 \\ a_2 \\ \ldots \\ a_m\end{pmatrix}\hat{u} &= \begin{pmatrix}a_1 & a_2 & \ldots & a_m\end{pmatrix}\begin{pmatrix}b_1 \\ b_2 \\ \ldots \\ b_m\end{pmatrix} \\
\\
(a_1^2+a_2^2+\ldots+a_m^2) \hat{u} &= a_1b_1+a_2b_2+\ldots+a_mb_m
\end{align}
$$

ゆえに微分で求めた解と同様に以下の解を得る．  

$$
\hat{u} = \frac{a_1b_1+a_2b_2+\ldots+a_mb_m}{a_1^2+a_2^2+\ldots+a_m^2}
$$

## 2.3.2
$Au = b$が解$u$を持つとすると，$\hat{u} = u$であることを示せ．$\hat{u} = u$が一意であるのはいかなる場合か？
### 解答

$$
A = \begin{pmatrix}\boldsymbol{a_1} & \boldsymbol{a_2} & \ldots & \boldsymbol{a_n}\end{pmatrix}
$$

以上のように行列$A$を考えると，$Au = b$が解$u$を持つとき，  
$b$は以下のように行列$A$の列ベクトルの組み合わせで表現できる．

$$
b = u_1\boldsymbol{a_1}+u_2\boldsymbol{a_2}+\ldots+u_n\boldsymbol{a_n}
$$

つまり，これは行列$A$の張る空間にベクトル$b$が含まれているということであり，  
射影$p$を与える$\hat{u}$は，このときの$u$と一致する． 

さらに$\hat{u}$が一意であることは，以下の式を満たす$u_1, u_2, \ldots, u_n$が一意であることと同値であり，  
列ベクトル$\boldsymbol{a_1}, \boldsymbol{a_2},\ldots, \boldsymbol{a_n}$が線形独立であるとき，$\hat{u}$が一意となる．

$$
b = u_1\boldsymbol{a_1}+u_2\boldsymbol{a_2}+\ldots+u_n\boldsymbol{a_n}
$$

## 2.3.7
$x = (0,1,2,3)$の点に対応して，$b = (4,1,0,1)$であるとき，最近接直線$C+Dx$の係数$\hat{u}=(C,D)$に対する正規方程式を立て，それを解け．各点が一つの直線上にあるならば，方程式が可解となるように，4個の方程式$Au = b$から始めよ． 
### 解答

この条件を行列で表すと

$$
\begin{pmatrix}1 & 0 \\ 1 & 1 \\ 1 & 2 \\ 1 & 3\end{pmatrix} \begin{pmatrix}C \\ D\end{pmatrix} = \begin{pmatrix}4 \\ 1 \\ 0 \\ 1\end{pmatrix}
$$

このとき

$$
A = \begin{pmatrix}1 & 0 \\ 1 & 1 \\ 1 & 2 \\ 1 & 3\end{pmatrix}，b = \begin{pmatrix}4 \\ 1 \\ 0 \\ 1\end{pmatrix}
$$

また $A^TA\hat{u} = A^Tb$ から$\hat{u} = (A^TA)^{-1}A^Tb$と計算できる．

```python
A = np.array([[1,0], [1,1], [1,2], [1,3]])
b = np.array([[4],[1],[0],[1]])

A_t = np.transpose(A)
m = np.dot(A_t, A)
m_inv = np.linalg.inv(m)

n = np.dot(A_t, b)
hat_u = np.dot(m_inv, n)

print(hat_u)
```

以上の結果から$C = 3, D = -1$であることがわかった．  
そこで実際に可視化をしてみる．

```python
x = np.array([0,1,2,3])
y = np.array([4,1,0,1])
y_line = 3 - x

plt.scatter(x,y)
plt.plot(x, y_line)
plt.show()
```

## 2.3.9
(微積分による問題7)$E=||b- Au||^2$を，4個の平方 --その最後の項は$(1-C-3D)^2$ --の和に書き下せ．導関数に関する方程式$\partial E/\partial C = 0$および$\partial E/\partial D = 0$を求めよ．それを2で除して，正規方程式$A^TA\hat{u} = A^Tb$を導け．
### 解答

$$E = (4-C)^2+(1-C-D)^2+(0-C-2D)^2+(1-C-3D)^2$$

この式の偏微分を求めると

$$
\begin{align}
\partial{E}/\partial{C} &= -2(4-C)-2(1-C-D)-2(-C-2D)-2(1-C-3D) = 0 \\
\partial{E}/\partial{D} &= -2(1-C-D)-4(-C-2D)-6(1-C-3D) = 0
\end{align}
$$

これらの連立方程式を整理すると，

$$
\begin{align}
4C + 6D &= 6 \\
6C + 14D &= 4
\end{align}
$$

ここで$A^TA\hat{u} = A^Tb$を振り返ると

$$
\begin{pmatrix}4 & 6 \\ 6 & 14\end{pmatrix}\begin{pmatrix}C \\ D\end{pmatrix} = \begin{pmatrix}6 \\ 4\end{pmatrix}
$$

これは最小二乗法から求めた連立方程式と一致している．

## 2.3.14
(推奨問題) $b = (b_1, \ldots, b_m)$を$a = (1, \ldots, 1)$を通る直線に射影しよう．1個の未知数の$m$個の方程式$au=b$を(最小二乗法で)解こう．

(a) $a^Ta\hat{u} = a^Tb$を解いて，$\hat{u}$は，$b$の平均であることを示せ．  
(b) $e-b-a\hat{u}$，分散$||e||^2$および標準偏差$||e||$を求めよ．  
(c) 水平線$\hat{b}=3$は$b=(1,2,6)$に最近接である．$p=(3,3,3)$は$e$に直交することを調べ，射影行列$P=A(A^TA)^{-1}A^T$を求めよ．
### 解答

(a)

$$
\begin{align}
(左辺) = a^Ta\hat{u} &= \begin{pmatrix}1 & 1 & \ldots & 1\end{pmatrix}\begin{pmatrix}1 \\ 1 \\ \vdots \\ 1\end{pmatrix}\hat{u} \\
&= m\hat{u}
\end{align}
$$

$$
\begin{align}
(右辺) = a^Tb &= \begin{pmatrix}1 & 1 & \ldots & 1\end{pmatrix}\begin{pmatrix}b_1 \\ b_2 \\ \vdots \\ b_m\end{pmatrix} \\
&= b_1 + b_2 + \ldots + b_m
\end{align}
$$

よって，

$$
\hat{u} = \frac{b_1 + b_2 + \ldots + b_m}{m}
$$

(b)
$$
\begin{align}
e &= \begin{pmatrix}b_1 \\ b_2 \\ \vdots \\ b_m\end{pmatrix} - \hat{u}\begin{pmatrix}1 \\ 1\\ \vdots \\ 1\end{pmatrix} = \begin{pmatrix}b_1-\hat{u} \\ b_2-\hat{u} \\ \vdots \\b_m-\hat{u}\end{pmatrix}\\
\end{align}
$$

よって分散$||e||^2$，標準偏差$||e||$は，

$$
\begin{align}
||e||^2 = e^Te = (b_1-\hat{u})^2+(b_2-\hat{u})^2+\ldots+(b_m-\hat{u})^2 \\
\\
||e|| = \sqrt{(b_1-\hat{u})^2+(b_2-\hat{u})^2+\ldots+(b_m-\hat{u})^2}
\end{align}
$$

(c)

$$
e = b - A\hat{u} = \begin{pmatrix}-2 \\ -1 \\ 3\end{pmatrix}
$$

ここで$e$と$p$について調べると，

$$
e \cdot p = -2 \times 3 -1 \times 3 + 3 \times 3 = 0
$$

ゆえに$e$と$p$は直交していることを再確認した．

また行列$A$は以下のような形をしているので，

$$
A = \begin{pmatrix}1 \\ 1 \\ 1\end{pmatrix}
$$

## 2.3.19
直線$b=C+Dt$が，$t=-1$で$b=7$，$t=1$で$b=7$，$t=2$で$b=21$を通るようにするときの3方程式を書き下せ．最小二乗解$\hat{u}=(C,D)$を求め，最近接直線を描け．
### 解答

3方程式を書き下すと，それぞれ代入して

$$
\begin{align}
7 &= C - D \\
7 &= C + D \\
21 &= C + 2D
\end{align}
$$

$$
A = \begin{pmatrix}1 & -1 \\ 1 & 1 \\ 1 & 2\end{pmatrix},
b = \begin{pmatrix}7 \\ 7 \\ 21\end{pmatrix}
$$

とおき，$A^TAu = A^Tb$を解くことによって最近接直線を求める．

```python
A = np.array([[1, -1], [1, 1], [1, 2]])
b = np.array([[7], [7], [21]])
A_t = np.transpose(A)

A_tA = np.dot(A_t, A)
A_tb = np.dot(A_t, b)
A_tA_inv = np.linalg.inv(A_tA)
hat_u = np.dot(A_tA_inv, A_tb)

print(hat_u)
```
